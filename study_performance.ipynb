```python
# # Function to create barplots that indicate percentage for each category.
def bar_perc(plot, feature):
    total = len(feature) # length of the column
    for p in plot.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total) # percentage of each class of the category
        x = p.get_x() + p.get_width() / 2 - 0.05 # width of the plot
        y = p.get_y() + p.get_height()           # hieght of the plot
        plot.annotate(percentage, (x, y), size = 12) # annotate the percentage
```

```python
no_of_male = len(dataset[dataset['gender'] == 'male'])
no_of_female = len(dataset[dataset['gender'] == 'female'])

plt.bar(['male','female'],[no_of_male,no_of_female],color = ['r','g'],label = ['male','female'])
plt.legend()
plt.show()
```

```python
#for education level
no_of_a = len(dataset[dataset['parental_level_of_education'] == "some college"])
no_of_b = len(dataset[dataset['parental_level_of_education'] == "associate's degree"])
no_of_c = len(dataset[dataset['parental_level_of_education'] == 'high school'])



plt.bar(["some college","associate's degree","high school"],[no_of_a,no_of_b,no_of_c], color = ['r','g','b','b'], label=["some college","associate's degree","high school",])
plt.title('No of people with Education level')
plt.legend()
plt.show()
```
